```python
fig = go.Figure()
fig.add_trace(go.Bar(x = [1,2,3,4,5],  y = [3,4,5,6,7]))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = tips.total_bill,  y = tips.tip, mode = "markers", marker_size = 12*tips['size']))
```

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook #Print graph in the notebook
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.iris import flowers
```

```python
flowers
```

```python
output_file('test.html')
p = figure(title = 'flower test')
p.xaxis.axis_label = 'petal_length'
p.yaxis.axis_label = 'petal_width'
p.circle(flowers['petal_length'], flowers['petal_width'])
show(p)
```

```python
#Graphs and it's use
#Scatter Plot --> relationship between two variable
#Line Chart --> on x-axis, we have time component
#Bar chart-> for categorical data
#count plot -> for categorical data
#pie chart -> for the distribution of different unique classes
#Box plot and voiline plot -> determine the outlier
#histogram and distribution plot -> for continuous data
#heat map -> understand correlation
#Piarplot -> all numerical replationship in one plot
#join plot -> relationship and distributuion of data
#swarmplot / Relplot ->scatterplot
#Lmplot -> rgression line
```
